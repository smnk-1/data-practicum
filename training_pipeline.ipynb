```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
X_train = pd.read_csv('data/X_train_original.csv')
y_train = pd.read_csv('data/y_train_original.csv')
print(X_train.shape, y_train.shape)
```

```python
y_train.describe()
```

```python
df = pd.concat([X_train, pd.Series(y_train.values.ravel(), name='y')], axis=1)
plt.figure(figsize=(12, 8))
for i, col in enumerate(X_train.columns, 1):
    plt.subplot(2, 3, i)
    sns.scatterplot(x=col, y='y', data=df, alpha=0.6)
    plt.title(f'{col} vs Target')
plt.tight_layout()
```

```python
corr_with_target = df.corr()['y'].drop('y').sort_values(ascending=False)

plt.figure(figsize=(5,4))
sns.heatmap(corr_with_target.to_frame(), cmap='coolwarm', annot=True, fmt=".3f")
plt.title('Корреляция признаков с целевой переменной')
```
