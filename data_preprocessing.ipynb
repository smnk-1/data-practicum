# **ЗАДАЧА**

Необходимо построить модель, которая по параметрам работы магической установки сможет предсказывать значение показателя стабильности системы — *“Гармония Бессмертия”*. Целевая переменная непрерывная, изменяется в пределах от 0 до 1, имеет тип float. Таким образом наша задача - задача **регрессии**.

| Название формулы                   | Что означает                                                                 | Формула                                   |
|-----------------------------------|----------------------------------------------------------------------------|-------------------------------------------|
| Общая мощность потоков             | Суммарная мощность двух потоков системы                                      | P_общ = Сила_Левого_Потока + Сила_Правого_Потока |
| Суммарная сила всех потоков        | Полная мощность системы с учётом притока                                      | P_сум = Сила_Левого_Потока + Сила_Правого_Потока + Приток_Силы_Потока |
| Общая сила ядра                    | Мощность ядра определяется ритмом работы и притоком энергии                  | P_ядра = Ритм_Магического_Ядра * Приток_Силы_Потока |
| Общее давление на выходе           | Давление потока на выходе системы                                             | P_вых = Приток_Давления_Выдоха_Истока + Давление_Выдоха_Истока |
| Магическая производительность      | Эффективность преобразования топлива в энергию                                | η_маг = Скорость_Перехода_Через_Портал / Эмульсия_Истока |
| Эффективность ядра                 | Насколько эффективно ядро преобразует поданное топливо                        | E_ядра = Сила_Ядра / Эмульсия_Истока |
| Магическая мощность установки       | Зависимость мощности установки от топлива, давления и перепада температур    | P_маг = f(Эмульсия_Истока, P_общ, ΔT), где ΔT = Пламя_Стихий - Температура_Вдоха_Истока |
| Износ источников (-)                   | Коэффициент деградации оборудования                                           | I_износа = Дыхание_Истока / Гармония_Бессмертия |
| Абсолютная разница стабильности (-)    | Величина отклонения состояния ядра от целевого уровня                          | ΔS = abs(Дыхание_Истока - Гармония_Бессмертия) |
| Нормированная стабильность  (-)       | Нормализованная мера нестабильности по скорости потока                        | S_норм = abs(Дыхание_Истока - Гармония_Бессмертия) / Скорость_Перехода_Через_Портал |


Некоторые формулы отмечены (-) т.к. они содержат целевую переменную и не могут быть использованы.

### Статистический подход

Для каждого признака с пропусками определим пул значений, которыми можно заполнить этот столбец, для каждого значения пула построим гистограммы до заполнения пропусков и после заполнения этим значением, вычислим значение KL-дивергенции, на основе данной информации отберем значение, которым можно заполнить пропуск

```python
X_train['Скорость перехода через портал'].describe()
```

Значения для тестового заполнения: срденее, медиана, мода

```python
values = {
    'mean': X_train['Скорость перехода через портал'].mean(),
    'median': X_train['Скорость перехода через портал'].median(),
    'mode' : X_train['Скорость перехода через портал'].mode()[0]
}
values
```

```python
from scipy.stats import entropy
```

```python
data_before = X_train['Скорость перехода через портал'].dropna()

hist_P, bins = np.histogram(data_before, bins=9, density=True)
hist_P = hist_P / hist_P.sum()

print("Гистограмма P (до заполнения) построена.")

plt.figure(figsize=(10, 4))
plt.hist(data_before, bins=bins, density=True, alpha=0.7, label='До заполнения', color='blue')
plt.title(f'Гистограмма {'Скорость перехода через портал'} до заполнения')
plt.xlabel('Скорость перехода через портал')
plt.ylabel('Плотность')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

kl_results = {}

for method, value in values.items():
    filled_data = X_train['Скорость перехода через портал'].fillna(value)

    hist_Q, _ = np.histogram(filled_data, bins=bins, density=True)
    hist_Q = hist_Q / hist_Q.sum()

    hist_Q_safe = np.where(hist_Q == 0, np.finfo(float).eps, hist_Q)
    hist_P_safe = np.where(hist_P == 0, np.finfo(float).eps, hist_P)

    kl_div = entropy(hist_P_safe, hist_Q_safe)
    kl_results[method] = kl_div
    plt.figure(figsize=(10, 4))
    plt.hist(data_before, bins=bins, density=True, alpha=0.7, label='До заполнения', color='blue')
    plt.hist(filled_data, bins=bins, density=True, alpha=0.5, label=f'После заполнения ({method})', color='orange')
    plt.title(f'Гистограмма {'Скорость перехода через портал'}\nЗаполнение: {method}, KL-дивергенция: {kl_div:.6f}')
    plt.xlabel('Скорость перехода через портал')
    plt.ylabel('Плотность')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    print(f"KL-дивергенция для {method} ({value:.5f}): {kl_div:.6f}")

best_method = min(kl_results, key=kl_results.get)
best_value = values[best_method]

print(f"\nЛучший метод: {best_method}, значение: {best_value:.5f}")
print(f"Минимальная KL-дивергенция: {kl_results[best_method]:.6f}")
```

### Машинный подход

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import entropy
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer, KNNImputer

X_train_numeric = X_train.select_dtypes(include=[np.number])

# IterativeImputer
imputer_iterative = IterativeImputer(random_state=42)
X_train_iterative = imputer_iterative.fit_transform(X_train_numeric)
X_train_iterative_df = pd.DataFrame(X_train_iterative, columns=X_train_numeric.columns, index=X_train_numeric.index)

# KNNImputer
imputer_knn = KNNImputer(n_neighbors=5)
X_train_knn = imputer_knn.fit_transform(X_train_numeric)
X_train_knn_df = pd.DataFrame(X_train_knn, columns=X_train_numeric.columns, index=X_train_numeric.index)

column_name = 'Скорость перехода через портал'

data_before = X_train[column_name].dropna()

hist_P, bins = np.histogram(data_before, bins=9, density=True)
hist_P = hist_P / hist_P.sum()

print("Гистограмма P (до заполнения) построена.")

methods = {
    'IterativeImputer (MICE)': X_train_iterative_df[column_name],
    'KNNImputer': X_train_knn_df[column_name]
}

for method_name, filled_series in methods.items():
    # Построим гистограмму (Q)
    hist_Q, _ = np.histogram(filled_series, bins=bins, density=True)
    hist_Q = hist_Q / hist_Q.sum()

    # Вычисляем KL-дивергенцию: D(P || Q)
    hist_Q_safe = np.where(hist_Q == 0, np.finfo(float).eps, hist_Q)
    hist_P_safe = np.where(hist_P == 0, np.finfo(float).eps, hist_P)

    kl_div = entropy(hist_P_safe, hist_Q_safe)
    print(f"KL-дивергенция для {method_name}: {kl_div:.6f}")

    plt.figure(figsize=(10, 4))
    plt.hist(data_before, bins=bins, density=True, alpha=0.7, label='До заполнения', color='blue')
    plt.hist(filled_series, bins=bins, density=True, alpha=0.5, label=f'После заполнения ({method_name})', color='orange')
    plt.title(f'Гистограмма {column_name}\nМетод: {method_name}, KL-дивергенция: {kl_div:.6f}')
    plt.xlabel(column_name)
    plt.ylabel('Плотность')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
```

## Устранение выбросов

```python
for column in X_train_knn_df.columns:
    plt.figure()
    plt.boxplot(X_train_knn_df[column])
    plt.title(column)
    plt.show
```

```python
outliers_columns = ['Печать Чародея']
for col in outliers_columns:
    plt.figure()
    plt.boxplot(X_train_knn_df[col])
    plt.show()
```

```python
def detect_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Возвращаем маску (True для выбросов)
    return (series < lower_bound) | (series > upper_bound)
```

```python
def detect_outliers_robust_zscore(series, threshold=3):
    median = series.median()
    mad = (series - median).abs().median()
    modified_z_score = 0.6745 * (series - median) / mad
    return modified_z_score.abs() > threshold
```

```python
for col in outliers_columns:
    outliers_iqr = detect_outliers_iqr(X_train_knn_df[col])
    outliers_rz = detect_outliers_robust_zscore(X_train_knn_df[col], threshold=3) # часто в источниках бывает 3.5, при таком значении порога 0 выбросов

    print(f"Количество выбросов по IQR в столбце '{col}': {outliers_iqr.sum()}")
    print(f"Количество выбросов по Robust Z-score в столбце '{col}': {outliers_rz.sum()}")
    outlier_intersection = outliers_iqr & outliers_rz
    print(f"Количество строк, являющихся выбросами по обоим методам: {outlier_intersection.sum()}")
    print(outlier_intersection)
```

```python
mask_keep_after_outlier_removal = ~outlier_intersection
X_train_after_outlier_removal = X_train_knn_df[mask_keep_after_outlier_removal]

initial_count = X_train_knn_df.shape[0]
final_count = X_train_after_outlier_removal.shape[0]
removed_count = initial_count - final_count

print(f"Размер датафрейма до удаления выбросов: {initial_count}")
print(f"Размер датафрейма после удаления выбросов: {final_count}")
print(f"Удалено строк: {removed_count}")

plt.figure(figsize=(6, 4))
plt.bar(['Остались', 'Удалены (Выбросы)'], [final_count, removed_count])
plt.title('Количество строк после удаления выбросов')
plt.ylabel('Количество строк')
plt.show()
```

## Устранение аномалий

```python
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler
# from pyod.models import DBSCAN as PyOD_DBSCAN
from sklearn.decomposition import PCA
```

```python
X_for_anomaly_detection = X_train_after_outlier_removal.select_dtypes(include=[np.number])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_for_anomaly_detection)
```

```python
contamination_rate = 0.02 # 2%

iso_forest = IsolationForest(contamination=contamination_rate, random_state=42)
oc_svm = OneClassSVM(nu=contamination_rate)
# dbscan = PyOD_DBSCAN(contamination=contamination_rate)

y_iso = iso_forest.fit_predict(X_scaled)
y_ocsvm = oc_svm.fit_predict(X_scaled)
# y_dbscan = dbscan.fit_predict(X_scaled)

anomaly_iso = y_iso == -1
anomaly_ocsvm = y_ocsvm == -1
# anomaly_dbscan = y_dbscan == -1

print(f"Isolation Forest нашёл аномалий: {anomaly_iso.sum()}")
print(f"One-class SVM нашёл аномалий: {anomaly_ocsvm.sum()}")
# print(f"DBSCAN нашёл аномалий: {anomaly_dbscan.sum()}")
```

```python
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled) # X_scaled - масштабированный numpy array

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# algorithms = [('Isolation Forest', anomaly_iso), ('One-class SVM', anomaly_ocsvm), ('DBSCAN', anomaly_dbscan)]
algorithms = [('Isolation Forest', anomaly_iso), ('One-class SVM', anomaly_ocsvm)]

for i, (name, anomaly_mask) in enumerate(algorithms):
    scatter = axes[i].scatter(X_pca[:, 0], X_pca[:, 1], c=anomaly_mask.astype(int), cmap='viridis', alpha=0.6)
    axes[i].set_title(name)
    axes[i].set_xlabel('PCA Component 1')
    axes[i].set_ylabel('PCA Component 2')
    handles, _ = scatter.legend_elements()
    axes[i].legend(handles, ["Normal", "Anomaly"], title="Class")

plt.tight_layout()
plt.show()
```
